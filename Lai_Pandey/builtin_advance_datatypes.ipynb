# 1. List 
List as name suggests is used to make a collection of any type or various type of data 
for example- 
1. list of names student 
2. list of subject and their scores 

```python
# if I have to write a program where you need to store the data of all the students in a class, how will you do that?
# for example if the class has only 50 students limit, then you can make 50 variables and store each student data

# Array - Array is also a way to store collection of data at one place, but the catch with array is the data fundamental datatype of the array is going to be same for all the data in that array.
# if you are making an array of number, then it is required to store only numbers in that array, you cannot store any string, boolean in that array, similarly, if the the array is of type string, you can only store string in that array but not the number or boolean.

# but when it comes to list, there is no such limitation, you can put any datatype in the list, and it will work.

# also in the traditional array, the size of the array is fixed while making the array, means you have to tell that array will have max limit of 50 student or 100 student

# but in case of list there is no way and there is no need as well to define the max size of the list, you can keep on entering data in the list and the list will keep on increasing. 

lst = ["Lai", "Anurag", "Abhinav", "Shashank", "Ayushi"]
# in programming we start counting from 0, means our first number is, not 1

print("Reading list data one by one - ")  # give me the 0th index data of the list `lst`

print(lst[0])
print(lst[1])
print(lst[2])
print(lst[3])
print(lst[4])
```
