```python
# Question K
df_of_PA_temp = df1[df1["STATE"] == "PA"]
data = dict(
    STATE = df_of_PA_temp["STATE"],
    COUNTRY = df_of_PA_temp["COUNTY"],
    V_PCT_12YRS = df_of_PA_temp["V_PCT_12YRS"]
)

df_of_PA = pd.DataFrame(data)
df_of_PA.sort_values(by=["V_PCT_12YRS"],inplace=True, ascending=False)


print("Top 5 countries in PA which has the highest V_PCT_12YRS")
highest_5_V_PCT_12YRS = df_of_PA["COUNTRY"].iloc[:5]
print(highest_5_V_PCT_12YRS)

print("\nTop 5 countries in PA which has the lowest V_PCT_12YRS")
lowest_5_V_PCT_12YRS = df_of_PA["COUNTRY"].iloc[-5:]
print(lowest_5_V_PCT_12YRS)
```
