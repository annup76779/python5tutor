## bubble sort
  - in this sorting algo we compare each and every element of the array and the smallest of them all is placed at the current pointer
``

```python
def bubble_sort(arr):
    """
    This algo/ function must work on number but in some cases may work on string.
    """
    n = len(arr)  # get the length of the array

    for i in range(n - 1):
        for j in range(0, n - i - 1):

            if arr[j] > arr[j + 1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

            print(arr, arr[j], arr[j+1])
        print(" -> ")

arr = [3,324,32,235,325,234,23,5342,45,346,54,645]
bubble_sort(arr) # sorting by bubble algo
print(arr)
```

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]  # the current element to be inserted at the currect position
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j+1] = arr[j]
            j -= 1
            print(arr, arr[j], arr[j+1])
        arr[j+1] = key
        print(arr, arr[j], arr[j+1])

        print("->")

arr = [3,324,2,235,325,234,23,5342,45,346,54,645]
insertion_sort(arr)
print(arr)
```
